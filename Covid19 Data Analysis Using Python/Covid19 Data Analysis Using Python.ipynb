### Calcuting and plotting the change in cases

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

#### For China, Italy, India and Spain (Max change in cases)

```python
china_cases = corona_dataset_aggregated.loc["China"].diff().max()
italy_cases = corona_dataset_aggregated.loc["Italy"].diff().max()
india_cases = corona_dataset_aggregated.loc["India"].diff().max()
spain_cases = corona_dataset_aggregated.loc["Spain"].diff().max()
total_max_case_changes = pd.Series({
    "China" : china_cases,
    "Italy" : italy_cases,
    "India" : india_cases,
    "Spain" : spain_cases
})

total_max_case_changes
```

```python
# Using iloc method
max_infected_rates = []
for i in range(len(corona_dataset_aggregated)):
     max_infected_rates.append(corona_dataset_aggregated.iloc[i].diff().max())
corona_dataset_aggregated["Max Infected Rates"] = max_infected_rates
```

```python
#Using loc method
countries = list(corona_dataset_aggregated.index)
max_infected_rates = []
for country in countries:
     max_infected_rates.append(corona_dataset_aggregated.loc[country].diff().max())
corona_dataset_aggregated["Max Infected Rates"] = max_infected_rates
corona_dataset_aggregated
```

```python
# Keeping only Max Infection Rate
corona_data = pd.DataFrame(corona_dataset_aggregated["Max Infected Rates"])
corona_data
```
